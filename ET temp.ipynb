```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import time
from datetime import timedelta
from requests.adapters import HTTPAdapter
from selenium import webdriver

s = requests.Session()
#s.mount('http://', HTTPAdapter(max_retries=2))
s.mount('https://', HTTPAdapter(max_retries=2))

headers={
'Referer': 'https://www.moneycontrol.com',
'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.149 Safari/537.36'
}

archive_url = 'https://economictimes.indiatimes.com/archive.cms'

while True:
    response = None
    try:
        response = s.get(archive_url, headers=headers, timeout=20)
        break
    except requests.exceptions.RequestException as err:
        print(f'Caught {err}... Sleeping for 80 sec and then retrying...')
        time.sleep(80)
        continue        

archive = response.text
soup = BeautifulSoup(archive,'html.parser')
content = soup.find('span',{'class':'normtxt'})
links = content.find_all('a')

monthly_links = []

for month in links:
    try:
        link_half = month.get('href')
        link = 'https://economictimes.indiatimes.com'+link_half
        #print('Link: ',link)
        monthly_links.append(link)

    except AttributeError:
        link = 'N/A'


monthly_links
```

```python
daily_links = []

for month_link in monthly_links:
    driver = webdriver.Chrome('C:\\Users\\Akshay Kaushal\\Downloads\\chromedriver')
    #driver.maximize_window()
    driver.get(month_link)
    source = driver.page_source
    soup1 = BeautifulSoup(source,'html.parser')
    driver.close()
    
    days = soup1.find('span',{'class':'normtxt'})
    days_links = days.find_all('a')
    
    for day_link in days_links:
        href = day_link.get('href')
        daily_link = 'https://economictimes.indiatimes.com'+href
        daily_links.append(daily_link)
        print('Day: ', daily_link)

daily_links
```

```python
news_count = 0
news_articles = {}
times_saved = 0

for news_links in dailylinks:
    n = 0
    while True:
        response2 = None
        try:
            response2 = s.get(news_links, headers=headers, timeout=20)
            break
        except requests.exceptions.RequestException as err:
            print(f'Caught {err}... Sleeping for 80 sec and then retrying...')
            time.sleep(40)
            n+=1
            if n == 5:
                break
            else:
                continue
        
    all_links = response2.text
    soup2 = BeautifulSoup(all_links,'html.parser')
    news = soup2.find_all('ul',{'class':'content'})#,{'class':'normtxt'})#,{'id':'calender'})


    for i in range(len(news)):
        tags = news[i].find_all('a')
        for tag in tags:
            href_ = tag.get('href')
            href1 = 'https://economictimes.indiatimes.com'+href_
            #print(href1)
            #all_news_links.append(href1)

            if news_count % 20000 == 1:
                start_time = time.monotonic()    

            if news_count % 100 == 0:
                print('\n', news_count)

            while True:
                response3 = None
                try:
                    response3 = s.get(href1, headers=headers, timeout=20)
                    break
                except requests.exceptions.RequestException as err:
                    print(f'Caught {err}... Sleeping for 80 sec and then retrying...')
                    time.sleep(80)
                    continue         


            news_page = response3.text
            soup3 = BeautifulSoup(news_page,'html.parser')

            try:
                if soup3.find('h1',{'class':'artTitle'}).text:
                    article = soup3.find('h1',{'class':'artTitle'}).text
                    published = soup3.find('time').text#,{'class':'_3Mkg- byline'}).text
                    text = soup3.find('div',{'class':'artText'}).text
                    link = href1
                    print('Title: ', article)
                    print('Published: ', published)
                    print('Link: ', link)
                    print('News:\n', text)                                                

            except AttributeError:
                try:
                    if soup3.find('div',{'class':'publish_on'}).text:
                        article = soup3.find('h1',{'class':'artTitle'}).text
                        published = soup3.find('div',{'class':'publish_on'}).text#,{'data-plugin':'dateformat'}).text
                        text = soup3.find('div',{'class':'artText'}).text
                        link = href1
                        #print('Title: ', article)
                        #print('Published: ', Updated)
                        #print('Link: ', link)
                        #print('News:\n', text)

                except AttributeError:
                    print('Attribute Error\n\n\n')
                    article = ''
                    published = ''
                    link = ''
                    text = ''

            news_count+=1
            news_articles[news_count] = [article, published, link, text]
            #print('Title: ', article, '\nPublished: ', published, '\nLink: ', link, '\nArticle: ', text, '\n\n---\n')
            print('\nPublished: ', published)

            write = news_count % 20000

            if write == 0:          # Autosave scraped data to a csv file once every 5k news aricles
                times_saved+=1
                end_time = time.monotonic()
                print(timedelta(seconds=end_time - start_time))
                print('\n')

                try:
                    news_df = pd.DataFrame.from_dict(news_articles,orient='index',columns=['Title','Published','Link','News'])
                    print(len(news_df))
                    print('\n')
                    news_df['Title'] = news_df['Title'].map(lambda x: x.encode('unicode-escape').decode('utf-8'))
                    news_df['Published'] = news_df['Published'].map(lambda x: x.encode('unicode-escape').decode('utf-8'))
                    news_df['Link'] = news_df['Link'].map(lambda x: x.encode('unicode-escape').decode('utf-8'))
                    news_df['News'] = news_df['News'].map(lambda x: x.encode('unicode-escape').decode('utf-8'))
                    #len(news_df)
                    news_df.to_csv('ET 0'+str(times_saved)+'.csv')       # Rename the file here
                    news_count = 0
                    news_articles = {}
                    time.sleep(5)

                except:
                    print('')
```
